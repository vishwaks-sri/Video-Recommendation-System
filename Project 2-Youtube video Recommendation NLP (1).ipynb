```python
data.drop(['Unnamed: 0'], axis='columns', inplace=True)
data
```

```python
data["publishedAt"] = pd.to_datetime(data["publishedAt"])
data.head(10)['description']
```

```python
# based on the description we shall find the similarity among the movies.
from sklearn.feature_extraction.text import TfidfVectorizer

#Define a TF-IDF Vectorizer Object. Remove all english stop words such as 'the', 'a'
tfv = TfidfVectorizer(min_df=3, max_features=None, 
                      strip_accents='unicode', analyzer='word', token_pattern=r'\w{1,}',
                      ngram_range=(1,3), 
                      stop_words='english')
```

```python
#Replace NaN with an empty string
data['description'] = data['description'].fillna('')
data.info()
```

```python
#Construct the required TF-IDF matrix by fitting and transforming the data
tfv_matrix = tfv.fit_transform(data['description'])
```

```python
tfv_matrix
```

```python
#Output the shape of tfidf_matrix
tfv_matrix.shape
```

```python
# Import sigmoid_kernel
from sklearn.metrics.pairwise import sigmoid_kernel

# Compute the sigmoid similarity matrix
sig = sigmoid_kernel(tfv_matrix, tfv_matrix)
```

```python
#Construct a reverse map of indices and movie titles
indices = pd.Series(data.index, index=data['v_title']).drop_duplicates()
```

```python
# Function that takes in video title as input and outputs the similar videos
def get_recommendation(title, sig=sig):
    # Get the index of the video that matches the title
    idx = indices[title]
    
    # Get the pairwsie similarity scores of all videos with that video
    sig_scores = list(enumerate(sig[idx]))
    
    # Sort the videos based on the similarity scores
    sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True)
    
    # Get the scores of the 10 most similar videos
    sig_scores = sig_scores[1:11]
    
    # Get the video indices
    video_indices = [i[0] for i in sig_scores]
    
    # Return the top 10 most similar videos
    return data['v_title'].iloc[video_indices]
```
