# EDA and PREPROCESSING

glob() function helps to read the files in same directory

```python
import pandas as pd
from glob import glob
import numpy as np
```

```python
stock_files = glob(r'F:\DATA SCIENCE\Project 2\youtube\data_*.csv')
stock_files
```

```python
data = pd.concat( (pd.read_csv(file) for file in stock_files), ignore_index = True)
data.head()
```

```python
data.columns
```

```python
data.drop(['Unnamed: 0','favoriteCount'], axis='columns', inplace=True)
print('Shape of Dataframe = ', data.shape)
data.head()
```

## Remove Duplicates

We can check for any duplicate values in the dataframe using function .duplicated() and remove them accordingly using function drop_duplicates().

```python
print("There are {} duplicated values.".format(data.duplicated().sum()))
data[data.duplicated(keep=False)].head(10)
```

```python
# drop duplicates except for the first occurrence
data = data.drop_duplicates()
data
```

```python
# check for the info after removing duplicates
print("There are {} duplicated values.\n".format(data.duplicated().sum()))
data.info()
```

## Finding and Replacing Nan values

```python
# check for the Nan values
data.isna().sum()
```

Our data set contains Nan values,
1. All numeric values are replaced with zero
2. All object values replaced with empty

```python
data['likeCount'] = data['likeCount'].fillna(0)
data['dislikeCount'] = data['dislikeCount'].fillna(0)
data['commentCount'] = data['commentCount'].fillna(0)
data['description'] = data['description'].fillna('')
```

```python
data.head()
```

```python
# check for the Nan values replacing nan values
data.isna().sum()
```

## Pre-Processing

From the info() function above, we can see that all of the columns of the DataFrame are in object format. We need to convert those column format into the correct format before we can perform analysis and visualization. For example, we need to convert the columns for "commentCount", "dislikeCount", "favoriteCount", "likeCount", "viewCount" into numeric data format. For "publishedAt" column, we need to convert it into datetime format. After that, we will sort the DataFrame by viewCount and likeCount for easy comprehension at a glance.

```python
# convert string into numeric and datetime columns
# sort table by viewCount and likeCount
from sklearn.utils import shuffle
numeric_columns = ["commentCount", "dislikeCount", "likeCount", "viewCount"]
data[numeric_columns] = data[numeric_columns].apply(pd.to_numeric)
data["publishedAt"] = pd.to_datetime(data["publishedAt"])
data_final = data.sort_values(by=['viewCount', 'likeCount'], ascending=False).reset_index(drop=True)
yt_data_final = shuffle(data_final).reset_index(drop=True)
yt_data_final.head()
```
